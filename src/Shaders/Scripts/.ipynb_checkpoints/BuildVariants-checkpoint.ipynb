```python
import os
import numpy as np
```

```python
def loadIncludes(codeFile):
    with open(codeFile, 'r') as f:
        code = f.readlines()
    outCode = []
    for line in code:
        if line.find('#include') > -1:
            _,fn = line.split(' ')
            fn = fn.replace('\n', '')
            fn = fn.replace('"', '')
            fn = '../GLSL/' + fn
            if os.path.isfile(fn):
                with open('../GLSL/' + fn, 'r') as f:
                    for includeLine in f.readlines():
                        outCode.append(includeLine)
            else:
                outCode.append(line)
        else:
            outCode.append(line)
    return outCode
```

```python
variants = [[""], ["Sheen"], ["Clearcoat"], ["Transmission"], ["Specular"], 
            ["Iridescence"], ["Anisotropy"], ["Translucency"],
            ["Sheen", "Specular"],
            ["Clearcoat", "Anisotropy"],
            ["Clearcoat", "Iridescence"],
            ["Clearcoat", "Transmission"],
            ["Transmission", "Specular"],
            ["Transmission", "Iridescence"]
           ]

vertCode = loadIncludes('../GLSL/DefaultGeneric.vert')
fragCode = loadIncludes('../GLSL/DefaultGeneric.frag')

matList = []
for r,d,f in os.walk('../GLSL/Materials'):
    for fn in f:
        fullPath = r + '/' + fn
        fullPath = fullPath.replace('\\', '/')
        if fn.endswith('glsl') and fn != 'UnityMaterialTemplate.glsl':
            matList.append((fn, fullPath))

outPath = '../GLSL/Generated'

for mat in matList:
    name = mat[0].split('.')[0]
    with open(mat[1], 'r') as f:
        matCode = f.readlines()
    if name == 'Default':
        for variant in variants:
            #variant = variants[0]
            variantName = name + "".join(variant)
            with open(outPath + '/' + variantName + '.vert', 'w') as outFile:
                for l in vertCode:
                    if l[0:8] == '#include':
                        outFile.writelines(matCode)
                    else:                 
                        outFile.write(l)
            with open(outPath + '/' + variantName + '.frag', 'w') as outFile:
                for i, l in enumerate(fragCode):
                    if i == 1 and 'Transmission' not in variant:
                        outFile.write('#define OPAQUE\n')
                    if l[0:8] == '#include':
                        for v in variant:
                            if v != "":
                                outFile.write('#define ' + v.upper() + "\n")
                        outFile.writelines(matCode)
                    else: 
                        outFile.write(l)
    else:
        if name == 'UnityRain_Refraction' or name == 'UnityGlass':
            name += 'Transmission'
        with open(outPath + '/' + name + '.vert', 'w') as outFile:
            for l in vertCode:
                if l[0:8] == '#include':
                    outFile.write('#define VERTEXSTAGE\n')
                    outFile.writelines(matCode)
                else:                 
                    outFile.write(l)
        with open(outPath + '/' + name + '.frag', 'w') as outFile:
            for l in fragCode:
                if l[0:8] == '#include':
                    outFile.writelines(matCode)
                else: 
                    outFile.write(l)
```

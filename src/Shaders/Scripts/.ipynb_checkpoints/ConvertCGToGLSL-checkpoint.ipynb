```python
def getUniforms(lines):
    uniforms = []
    codeLines = []
    subShader = False
    codeBlock = False
    for idx, l in enumerate(lines):
        if l.find('SubShader') > -1:
            subShader = True
        if subShader:
            if l.find('};') > -1:
                codeBlock = True
                continue
            if l.find('ENDCG') > -1:
                break
        if codeBlock:
            line = l
            line = line.replace('\t', '')
            line = line.replace('\n', '')
            if line.find('uniform') > -1:
                line = line.replace(';', '')
                uniforms.append((line.split(' ')[1], line.split(' ')[2]))
            else:
                codeLines.append(l)
    finalcolorFunc = None
    if lines[idx + 2].find('finalcolor') > -1:
        parts = lines[idx + 2].split(' ')
        for p in parts:
            if p.find('finalcolor') > -1:
                finalcolorFunc = p.split(':')[1]
        print('shader has final color func: %s' % finalcolorFunc)
    
    sortedUniforms = []
    for u in uniforms:
        if u[0] == 'float4' or u[0] == 'half4':
            sortedUniforms.append(('vec4', u[1]))
    for u in uniforms:
        if u[0] == 'float' or u[0] == 'half':
            sortedUniforms.append(('float', u[1]))
    for u in uniforms:
        if u[0] == 'sampler2D':
            sortedUniforms.append(('TextureInfo', u[1]))

    return sortedUniforms, codeLines, finalcolorFunc

def getTokens(line):
    whitespaces = [' ', '\t', '\n', '\r']
    operators = ['+', '-', '*', '/', '=', '(', ')', ',', '.', ';']
    tokens = []
    currentToken = ""
    currentIndex = 0
    for idx, chr in enumerate(line):
        if chr not in whitespaces and chr not in operators:
            currentToken += chr
        else:
            if currentToken != "":
                tokens.append((currentToken, currentIndex))
            currentToken = ""
            currentIndex = idx
    return tokens

def convertToGLSL(codeLines, uniforms):
    
    replaceDict = {
        'float2': 'vec2',
        'float3': 'vec3',
        'float4': 'vec4',
        'half2' : 'vec2',
        'half3' : 'vec3', 
        'half4' : 'vec4',
        'sampler2D': 'TextureInfo',
        'frac(' : 'fract(', # TODO: maybe better to implement hlsl functions
        'fmod' : 'mod', # TODO: hlsl fmod has not the same behaviour as glsl mod!!!
        'lerp': 'mix',
        'ddx': 'dFdx',
        'ddy': 'dFdy',
        'inline': '',
        '(float)': '',
        '(int)': '',
        # WORKAROUNDS until we find a better way to solve this...
        'Refraction( i, o, _IndexofRefraction, _ChromaticAberration )' : 'vec3(Refraction( i, o, _IndexofRefraction, _ChromaticAberration ))',
        'i.screenPos * i.worldPos' : 'vec3(i.screenPos) * i.worldPos',
    }
    
    parsedCode = []
    for l in codeLines:
        for k in replaceDict.keys():
            if l.find(k) > -1:
                l = l.replace(k, replaceDict[k])
        c = l
        c = c.replace('\t', '')
        c = c.replace('\n', '')
        c = c.replace(';', '')
        parts = c.split(' ')
        if len(parts) == 4 and (parts[0] == 'vec2' or parts[0] == 'vec3'):
            typename, name, op, value = parts
            if value == '0':
                l = l.replace('0', typename + '(' + value + ')')
    
        if l.find('( int )') > -1 or l.find('(int)') > -1:
            l = l.replace('( int )', 'int(')
            l = l.replace(');', '));')
    
        # tokens = getTokens(l)
        # for u in uniforms:
        #     name = u[1]
        #     if name in tokens:
        #         count = tokens.count(name)
        #         l = l.replace(name, 'material.' + name, count)

        if len(l) > 3:
            parsedCode.append(l[2:])
        else:
            parsedCode.append(l)
            
    return parsedCode
```

```python
globals = {
    'INTERNAL_DATA' : '',
    'UNITY_MATRIX_V' : 'camera.view',
    'UNITY_INITIALIZE_OUTPUT(Input, Ouput)' : '',
    '_Time' : 'camera.time',
    '_ProjectionParams' : 'camera.projParams',
    #'material._GrabTexture' : 'grabTexture' # TODO: this is just workaround for screen grab tex
}

assetPath = 'C:/workspace/code/Archviz/Assets'
shaderList = []
for r,d,f in os.walk(assetPath):
    for fn in f:
        fullPath = r + '/' + fn
        fullPath = fullPath.replace('\\', '/')
        if fn.endswith('shader'):
            shaderList.append((fn, fullPath))

for shaderFile, shaderPath in shaderList:

    print('converting shader %s' % shaderFile)
    shaderName = shaderFile.split('.')[0]
    
    with open(shaderPath, 'r') as f:
        lines = f.readlines()
    
    uniforms, code, finalFunc = getUniforms(lines)
    glslCode = convertToGLSL(code, uniforms)
    
    with open('../GLSL/Materials/Unity/UnityMaterialTemplate.glsl', 'r') as f:
        lines = f.readlines()
    
    outlines = []
    for line in lines:
        if line.find('@UnityMaterial') > -1:
            for u in uniforms:
                s = u[0] + ' ' + u[1] + ';'
                outlines.append('\t' + u[0] + ' ' + u[1] + ';\n')
        elif line.find('@UnitySurface') > -1:
            for k,v in globals.items():
                outlines.append('#define ' + k + ' ' + v + '\n')
            if finalFunc != None:
                outlines.append('#define UNITY_PASS_FORWARDBASE\n')
                outlines.append('#define FINALCOLOR\n')
            outlines.append('\n')  
            for u in uniforms:
                if u[1] == '_GrabTexture':
                    outlines.append('#define _GrabTexture grabTexture\n')
                else:
                    outlines.append('#define ' + u[1] + ' material.' + u[1] + '\n')
            outlines.append('\n')
            for l in glslCode:
                outlines.append(l)

            outlines.append('Input unityInput;\n')
            if shaderName == 'Vegetation':
                outlines.append('SurfaceOutputStandardSpecular unitySurface;\n')
            else:
                outlines.append('SurfaceOutputStandard unitySurface;\n')
        else:
            outlines.append(line)
            
    if finalFunc != None: # adding final color func, according to unity doc it has always the same signature
        outlines.append('\n')
        outlines.append('vec4 finalColor(vec4 color)\n')
        outlines.append('{\n')
        outlines.append('\t' + finalFunc + '(unityInput, unitySurface, color);\n')
        outlines.append('\treturn color;\n')
        outlines.append('}\n')    
    
    with open('../GLSL/Materials/Unity/Unity' + shaderName + '.glsl', 'w') as f:
        f.writelines(outlines)
```
